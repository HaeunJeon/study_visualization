```python
from moviepy.editor import ImageSequenceClip, vfx
import os

# Path to the images uploaded by the user
image_paths = [
    "/Untitled.png",
    "/Untitled (1).png",
    "/Untitled (2).png",
    "/Untitled (3).png"
]

# Load the images
clips = []
for img_path in image_paths:
    # Load the image and add a zoom effect
    clip = (ImageClip(img_path)
            .set_duration(3)
            .resize(lambda t: 1 + 0.05 * t)  # Zoom-in effect
            .set_position('center'))
    clips.append(clip)

# Concatenate clips with crossfade effect
video = concatenate_videoclips(clips, method="compose")

# Write the result to a file
video_path = '/fashion_show.mp4'
video.write_videofile(video_path, fps=24)

video_path
```
